```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python

images_dir = 'images_gray'
test_images_dir = 'test'
train_images_dir = 'train'

# 1. Görüntüyü Yükle ve Gri Tonlamaya Çevir
gray_image =  cv2.imread(os.path.join(images_dir, train_images_dir,'train_easy_1.png'), cv2.IMREAD_GRAYSCALE)
```

```python
blurred_image = cv2.GaussianBlur(gray_image,(5,5),0)

#edge detection with sobel
sobelx = cv2.Sobel(blurred_image,cv2.CV_64F,1,0,ksize=5)
sobely = cv2.Sobel(blurred_image,cv2.CV_64F,0,1,ksize=5)
sobel = np.sqrt(sobelx**2 + sobely**2)

#edge detection with laplacian
laplacian = cv2.Laplacian(blurred_image,cv2.CV_64F)

#edge detection with canny
canny = cv2.Canny(blurred_image,100,200)

#plot
plt.figure(figsize=(20,20))
plt.subplot(221),plt.imshow(gray_image, cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(222),plt.imshow(sobel, cmap = 'gray')
plt.title('Sobel Image'), plt.xticks([]), plt.yticks([])
plt.subplot(223),plt.imshow(laplacian, cmap = 'gray')
plt.title('Laplacian Image'), plt.xticks([]), plt.yticks([])
plt.subplot(224),plt.imshow(canny, cmap = 'gray')
plt.title('Canny Image'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
blurred_image = cv2.GaussianBlur(gray_image,(5,5),0)

#edge detection with sobel
sobelx = cv2.Sobel(blurred_image,cv2.CV_64F,1,0,ksize=5)
sobely = cv2.Sobel(blurred_image,cv2.CV_64F,0,1,ksize=5)
sobel = np.abs(sobelx) + np.abs(sobely)

# Convert sobel to 8-bit image
sobel_8u = np.uint8(255 * sobel / np.max(sobel))

#edge detection with laplacian
laplacian = cv2.Laplacian(blurred_image,cv2.CV_64F)

#edge detection with canny
canny = cv2.Canny(sobel_8u,100,200)

#plot
plt.figure(figsize=(20,20))
plt.subplot(221),plt.imshow(gray_image, cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(222),plt.imshow(sobel, cmap = 'gray')
plt.title('Sobel Image'), plt.xticks([]), plt.yticks([])
plt.subplot(223),plt.imshow(laplacian, cmap = 'gray')
plt.title('Laplacian Image'), plt.xticks([]), plt.yticks([])
plt.subplot(224),plt.imshow(canny, cmap = 'gray')
plt.title('Canny Image'), plt.xticks([]), plt.yticks([])
plt.show()
```
