```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#generate a cos signal for 8 period and a sinus signal for 4 period then take the negative of the signal and plot them over 

w1 = 3*np.pi/2

n = 10000
x = np.linspace(0, 2*np.pi, n)
a = np.append(np.ones(n//2), -np.ones(n//2))
a_x = a * x
cos_signal = np.cos(w1*a_x)
sin_signal = -np.sin(w1*a_x)

plt.plot(x, cos_signal)
plt.plot(x, sin_signal)
# plot x axis at y = 0
plt.axhline(y=0, color='k')
#add legend
plt.legend(['cos signal', 'sin signal'])
plt.xlabel('x')
plt.ylabel('Amplitude')
plt.show()
```

```python
# Find indices where the cosine signal crosses zero on the falling edge
zero_crossings = np.where(np.diff(np.signbit(cos_signal)) & (cos_signal < 0))[0]

plt.figure(figsize=(10, 5))
plt.plot(x, cos_signal)
plt.plot(x, sin_signal)

# Plot vertical lines at zero crossings
for idx in zero_crossings:
    plt.axvline(x=x[idx], color='r', linestyle='--')

# Plot x axis at y = 0
plt.axhline(y=0, color='k')

# Add legend
plt.legend(['cos signal', 'sin signal', 'zero crossings'])
plt.xlabel('x')
plt.ylabel('Amplitude')
plt.show()
```
